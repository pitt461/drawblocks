```python
def getColor (occData):
    occData = float(occData)
    if occData <= 0.34:
        return 'green'
    elif occData > 0.34 and occData <= 0.68:
        return 'yellow'
    elif occData > 0.68 and occData <= 1:
        return 'red'
    elif occData > 1:
        return 'black'
```
